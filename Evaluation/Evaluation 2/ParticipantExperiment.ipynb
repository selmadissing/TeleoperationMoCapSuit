```python
import pandas as pd
from scipy.stats import wilcoxon
import numpy as np
import matplotlib.pyplot as plt

# Assuming df1 and df2 are your two dataframes
video = pd.read_csv("SurveyVideo.csv")
suit = pd.read_csv("SurveySuit.csv")

# Initialize a list to hold results
results = []

# Perform Wilcoxon tests as before, but this time only for Q2 and store the results in the list
for i in range(1, 6):
    task_question_df1 = suit[f"Task {i} - Q2"]
    task_question_df2 = video[f"Task {i} - Q2"]
    w, p = wilcoxon(task_question_df1, task_question_df2)
    results.append((-np.log10(p), f'Task {i} - Q2'))  # Store -log10(p) and task-question label

# Sort results by -log10(p) value
#results.sort()

# Separate the -log10(p) values and labels into separate lists
p_values, labels = zip(*results)

# Create the plot
plt.figure(figsize=(10, 8))  # Adjust as necessary
plt.barh(labels, p_values, color='skyblue')  # Horizontal bar plot
plt.xlabel('-log10(p-value)')
plt.title('Wilcoxon Signed-Rank Test Results for Q2')
plt.show()
```

```python
import pandas as pd
from scipy.stats import wilcoxon
from tabulate import tabulate

# Assuming df1 and df2 are your two dataframes
video = pd.read_csv("SurveyVideo.csv")
suit = pd.read_csv("SurveySuit.csv")

# Initialize lists to hold the results for each task
tasks = []
statistics = []
p_values = []

# Perform Wilcoxon tests as before, but this time only for Q2 and store the results in the lists
for i in range(1, 6):
    task_question_df1 = suit[f"Task {i} - Q2"]
    task_question_df2 = video[f"Task {i} - Q2"]
    w, p = wilcoxon(task_question_df1, task_question_df2)
    tasks.append(f'Task {i}')
    statistics.append(w)
    p_values.append(p)

# Create the combined table for all tasks
table_data = {'Task': tasks, 'Wilcoxon Statistic': statistics, 'p-value': p_values}
df = pd.DataFrame(table_data)
df.set_index('Task', inplace=True)

# Format the table for display
table = tabulate(df, headers='keys', tablefmt='fancy_grid')

# Display the table
print(table)


```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the tasks (we will only plot for Q2 of each task)
tasks = [f'Task {i} - Q2' for i in range(1, 6)]  # this will create ['Task 1 - Q2', 'Task 2 - Q2', ..., 'Task 5 - Q2']

# Calculate mean for each task for both systems
means_system1 = [suit[task].mean() for task in tasks]
means_system2 = [video[task].mean() for task in tasks]

x = np.arange(len(tasks))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()

# Generate bars for each system
rects1 = ax.bar(x - width/2, means_system1, width, label='System 1')
rects2 = ax.bar(x + width/2, means_system2, width, label='System 2')

# Add some text for labels, title, and custom x-axis tick labels, etc.
ax.set_xlabel('Tasks')
ax.set_ylabel('Mean Scores')
ax.set_title('Mean Scores per Task for Both Systems')
ax.set_xticks(x)
ax.set_xticklabels(tasks, rotation=45)
ax.legend()

# Function to auto-label the bars
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(round(height, 2)),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()  # To make sure labels fit into the plot area

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Define the tasks (we will only plot for Q2 of each task)
tasks = [f'Task {i} - Q2' for i in range(1, 6)]  # this will create ['Task 1 - Q2', 'Task 2 - Q2', ..., 'Task 5 - Q2']
task_names = [f'Task {i}' for i in range(1, 6)]  # Define your task names for titles

fig, axs = plt.subplots(1, len(tasks), figsize=(20, 6), sharey=True)

colors = ['#3498db', '#2ecc71']  # colors for the box plots, you can choose your own

# Generate boxplot for each task
for i, task in enumerate(tasks):
    data_to_plot = pd.DataFrame({'Suit': suit[task].dropna(), 'Video': video[task].dropna()})

    sns.boxplot(data=data_to_plot, ax=axs[i], palette=colors, width=0.5)
    
    axs[i].set_title(task_names[i])
 

plt.subplots_adjust(wspace=0.3)  # Adjust the space between plots
fig.suptitle("I am satisfied with the robot's movements when performing the task", fontsize=14, fontweight='bold')
plt.show()
```
